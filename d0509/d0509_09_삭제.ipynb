```python
import pandas as pd

df = pd.read_excel('stu_e.xlsx',index_col='지원번호')
df
```

```python
df.index
df.columns
```

```python
# 국어점수 80점 이상인 학생들 행 삭제
df['국어'] >= 80
df[df['국어'] >= 80].index
df.drop(index=df[df['국어'] >= 80].index,inplace=True)
df
```

```python
# columns 삭제 -> 정확한 컬럼명을 입력해야 삭제
# index 삭제 -> 정확한 index명을 입력해야 삭제

# index 1개 삭제
df.drop(index='1번',inplace=True)
df
```

```python
## 부분 수정
df.loc['5번','국어'] = 95
df
```

```python
# index 여러개 삭제
df.drop(index=['3번','5번'],inplace=True)
df
```

```python
# 조건에 맞는 사람들만 삭제
filt = df['과학'] > 70
df.drop(index=df[filt].index,inplace=True)
df
```

```python
# 수정
df['학교'] = df['학교'] + '등학교'
df
```

```python
# 수정
df['학교'].replace({'구로고등학교':'디지털고','단지고등학교':'영등포고'},inplace=True)
df
```
