dropna - Nan 처리: 지우기

axis : index,columns how : any,all

```python
# Nan 이 들어가있는 컬럼을 전체 삭제 (= 열삭제)
df.dropna(axis='columns', inplace=True)
df
```

```python
# Nan 들어가 있는 컬럼에서 모든 컬럼이 Nan인 것만 삭제
df.dropna(axis='columns',how='all',inplace=True)
df
```

```python
# Nan이 들어가 있는 컬럼에서 1개라도 Nan이 포함되어 있으면 삭제
df.dropna(axis='columns',how='any',inplace='True')
```

```python
# Nan 이 들어가있는 index에서 1개라도 Nan이 포함되어 있으면 삭제
df.dropna(axis='index',how='any',inplace=True) # 행 삭제
```
