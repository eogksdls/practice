```python
import pandas as pd

df = pd.read_excel('stu_e.xlsx', index_col='지원번호')

df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df.fillna('미선택',inplace=True)
# 공백을 '미선택'으로 채움
```

```python
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
#  문구를 추가해서 수정
df['학교'] = df['학교'] + '등학교'
df
```

```python
# 과학 점수를 +5점 더해서 출력하시오
df['과학'] = df['과학'] + 5
df
```

```python
# 국어 점수 +10점을 하시오
df['국어'] = df['국어'] + 10
df
```
