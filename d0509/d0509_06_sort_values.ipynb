```python
import pandas as pd
df = pd.read_excel('stu_e.xlsx',index_col='지원번호')
df
```

```python
# 키의 데이터를 가지고 정렬
df.sort_values('키',ascending=False)  # 키가 큰 순으로
# 키, 이름으로 정렬, 키가 동일한 경우 이름으로 정렬
df.sort_values(['키','이름'],ascending=False)
# 키는 역순정렬, 이름은 순차정렬
df.sort_values(['키','이름'],ascending=[False,True])
```

```python
import pandas as pd

df = pd.read_csv('stu_score1.csv')
df

# kor - 역순정렬, id - 순차정렬
df.sort_values(['kor','id'],ascending=[False,True])
```

```python
# 평균 75점 이상이면서, 평균 - 역순정렬, 이름 - 순차정렬
filt = df['avg'] >= 75

df[filt].sort_values(['avg','id'],ascending=[False,True])
```

```python
# id 대소문자 구분 없이 a가 포함된, 평균-역순정렬, 국어-순차정렬
filt = df['id'].str.lower().str.contains('a',na=False)
df[filt].sort_values(['avg','id'],ascending=[False,True])
```
