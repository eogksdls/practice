```python
import pandas as pd
df = pd.read_excel('stu_e.xlsx',index_col='지원번호')
df
```

```python
# 사회 컬럼을 삭제
# df.drop(columns=['사회'],inplace=True)
df['음악'] = 0
df
```

```python
# 음악점수는 과학점수가 80점이상이면 80, 미만이면 65점 입력
filt = df['과학'] >= 80
df.loc[filt,'음악'] = 80
df.loc[~filt,'음악'] = 65
df
```

```python
# 합계 컬럼을 추가해서 국어+영어+수학+과학+음악 으로 추가
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['음악']
df
```

```python
# 컬럼의 위치는 음악이 과학 바로 뒤로 컬럼 위치 변경
a_list = list(df.columns)
a_list
# type(a_list)
```
